```python
from transformers import pipeline
import numpy as np
import time

# 선택 분류
choose_create = pipeline("zero-shot-classification", model = "MoritzLaurer/mDeBERTa-v3-base-mnli-xnli")

# 스토리 생성
story_create = pipeline("text-generation", model = "EleutherAI/polyglot-ko-1.3b")

# 목표 지점 정하기
goal = np.random.randint(3, 20)

# 질문 개수 랜덤 생성
question_total = goal + 5

print("총 질문 개수:", question_total)
print("목표 거리 : ", goal)
print("게임 시작!\n")

questions = []

position = 0

# question_total 만큼 질문 생성
for i in range(question_total):

    print(f"문제 {i+1} 생성 중")
    story = story_create(
        f"당신은 게임 진행자입니다. 다음 조건에 맞는 문장 하나만 출력하세요.\n조건: 매우 짧고 긍정적이거나 부정적인 감정이 확실한 한국어 문장\n출력:",
        max_new_tokens=20
    )

    question = story[0]["generated_text"]
    questions.append(question)
    
    print("=================================================\n")

# 질문 진행
for i in range(question_total):

    print("=================================================\n")
    print("현재 위치 : ", position, " / ", goal)
    print("문제", i+1)
    print(questions[i])


    start = time.time()
    inputs = input("긍정 / 부정 입력 (5초 제한, 종료하려면 '제 수준이 미달입니다'): ")
    end = time.time()

    print(start - end)

    # 종료
    if inputs == "제 수준이 미달입니다":
        print("수듄ㅋ 🤭 풒ㅋ")
        break
    
    # 시간 초과
    if end - start > 5 :
        move = np.random.randint(0, 3)
        print("시간 초과지롱~~😛", move, "만큼 응~ 뒤로 가~")
        position = position - move

    else :
        result = choose_create(questions[i], candidate_labels = ["긍정", "부정"])
        correct = result["labels"][0]

        if inputs == correct :
            move = np.random.randint(0, 3)
            print("정답입니다", move, "칸 전진!")
            position = position + move

        else :
            move = np.random.randint(0, 3)
            print("이걸 틀려?", move, "뒤로 가라")
            position = position - move

    if position < 0 :
        position = 0

    if position >= goal :
        print("\n🎉 Goal 🎉")
        break

# 게임이 끝난 상황
if position < goal :
    print("게임 종료 Goal 지점에 도착 실패")
```
